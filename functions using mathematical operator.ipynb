```python
# even or odd checker
```

```python
def even_or_odd(num):
    if num%2==0:
        return "even"
    else:
        return "odd"
n=int(input("enter your number"))
print(even_or_odd(n))
```

```python
#find largest of 3 numbers
```

```python
def largest(a,b,c):
    if a>=b and a>=c:
        return a
    elif b>=a and b>=c:
        return b
    else:
        return c
print("largest:", largest(1,40,10))
```
