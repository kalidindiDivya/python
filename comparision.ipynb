```python
# find the largest of three numbers
```

```python
a=int(input("enter first number"))
b=int(input("enter seconf number"))
c=int(input("enter third number"))
if(a>=b and a>=c):
    print(f"a={a} is larger than {b} and {c}")
elif(b>=a and b>=c):
    print(f"b={b} is larger than {a} and {c}")
else:
    print(f"c={c} is larger than {a} and {b}")
```
