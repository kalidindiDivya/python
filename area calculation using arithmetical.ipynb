```python
we use float , because usually areas have decimals(5.1, 3.4 etc)
INTEGER is also we have to use ( if all the sides are whole numbers only), interger is a whole number(1, 10, -1 etc)
```

```python
side=float(input("enter the side of square"))
area= side* side
print("area of square", area)
```
