```python
#FILTER is used to pick out only the items that matches the condition
#syntax----> filter(function, collection(list , tupple etc))
#function returns true or false
# to see the result , wrap with list()
```

```python
# get only even number
```

```python
# traditional method
def is_even(x):
    if x%2==0:
        return True
    else:
        return False
List=[2,4,3,10,9,7,11]
filter(is_even, List) # it gives stored location address
list(filter(is_even, List)) # gives proper output , that what we want
```

```python
# using lambda function
List=[2,4,3,10,9,7,11]
even=filter( lambda x: x%2==0,List)
print(list(even))
```

```python
# using lambda function
List=[2,4,3,10,9,7,11]
odd=filter( lambda x: x%2!=0,List)
print(list(odd))
```

```python
# numbers >50
L=[20,10,50,80,40,61]
greater=filter(lambda x:x>50,L)

print(list(greater))
```

```python
# positive numbers
L= [-3,-2,-1,0,1,2,-5,4]
num=filter(lambda x:x>=0, L)
print(list(num))
```

```python
# words with > 3 letters
```

```python
N=["divya","dev","laxmi","mani"]
words=filter(lambda w: len(w)>3, N)
print(list(words))
```

```python
# MAP is a function which is used to generate every item ina list/collection 
# syntax----> map(function, collection(list, tuple etc))
# function , wt to do in each item
# to see the result , wrap with list()
```

```python
# traditional method
def is_even(x):
    if x%2==0:
        return True
    else:
        return False
List=[2,4,3,10,9,7,11]
map(is_even, List) # it gives stored location address
list(map(is_even, List)) # gives proper output , that what we want
```

```python
L=[2,4,3,10,9,7,11]
even=map(lambda x: x%2==0,L)
print(list(even))
```

```python
#square root
sqrt=map(lambda x:x**2,L)
print(list(sqrt))
```
