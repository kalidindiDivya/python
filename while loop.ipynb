```python
# while loop
while loop keep repeating a block of code as lost as a condition is true, once the condition become false, it stops
# syntax-----> while condition:
```

```python
count=1
while count<=5:
 print(count)
 count=count+1
```

```python
how it works
1. check if counr <=5---> yes, its 1
2. print 1, then add 1--->count becomes 2
3. check again---> yes its 2
4. repeat until count becomes 6
5. 6<=5, false ----> loop stops
```

```python
# find sum of numbers from 1 to 100
```

```python
num=1
total=0
while(num<=100):
    total=total+num
    num=num+1
print("sum=", total)
```
